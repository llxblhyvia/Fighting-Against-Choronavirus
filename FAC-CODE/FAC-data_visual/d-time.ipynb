```python
import pandas as pd
import numpy as np
import warnings
import re
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']='SimHei'
plt.rcParams['axes.unicode_minus'] = False  # 负号正常显示
```

```python
v_data = pd.read_csv("2021VAERSDATA-utf8.csv",encoding="utf-8")
v_data=v_data.sort_values(by='RECVDATE')
step=int(v_data.shape[0]/10)
l1,l2=0,step
key = ['headache','arm','fever','chills','swelling','nausea','rash']
fre=np.zeros([10,len(key)])
date=[]


for i in range(10):
    data=v_data.iloc[l1:l2,:]
    date.append(data.RECVDATE.tolist()[0])
    l1=l2
    l2+=step
    dic_v_data ={}
    content = data['SYMPTOM_TEXT'].tolist()
    len_c = len(content)
    try:
        for x in range(len_c):
            list1 = re.split('[ ?,.:();-@"]',content[x])
            for word in list1:
                if word not in dic_v_data:
                    dic_v_data[word] = 1
                else:
                    dic_v_data[word] += 1
    except:
        print('')
    fre[i]=np.array([dic_v_data[word] for word in key])
    
date=[x[:10] for x in date]
```

```python
df_fre=pd.DataFrame(fre,columns=key)
df_fre["date"]=date
df_fre=df_fre[["headache","date"]]

plt.figure(figsize=(14, 8))
sns.lineplot('date', 'headache', data=df_fre, label='headache',marker="o")

sns.despine()

plt.grid()
plt.legend(fancybox=True, prop={'size': 12})
plt.xlabel('日期',fontsize=18)
plt.ylabel('频率',fontsize=18)
plt.title('症状-头痛随时间推移出现频率的变化',fontsize=25)
plt.savefig('症状-头痛随时间推移出现频率的变化')
```
